##IMPORTING LIBRARIES


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

##LOAD THE DATASET

```python
df=pd.read_csv('/content/Housing.csv')
```

##DATA PREPROCESSING AND ANALYSIS

```python
df.drop_duplicates()
```

```python
df.dropna(how='any', subset=['area', 'price'], inplace=True)
```

```python
df
```

##Selecting Independent and Dependent Variable

```python
x=df.iloc[:,1:2] #dependent variable
y=df.iloc[:,:1] #independent variable
```

##Splitting the dataset

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=2)
```

##Model Fitting

```python
model=LinearRegression() #object 
model.fit(X_train,y_train)
```

```python
plt.scatter(X_train, y_train,color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.show()
```

```python
print(model.intercept_, model.coef_)
y_pred = model.predict(X_test) 
y_pred
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
y_test
```

```python
r2_score = model.score(X_train,y_train)
print(r2_score*100,'%')
```
